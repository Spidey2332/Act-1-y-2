#1- Regresión lineal

```python
#analizamos caso 1 de correlación entre variable de número de clientes en la tienda respecto al número de estantes en la tienda, el número de refrigeradores y el número de empleados permanentes
Vars_Indep= dflimpio[['288_number_app_purchases_month','81_number_online_purchases_month']]
Var_Dep= dflimpio['289_number_whatsapp_purchases_month']
```

```python
#se define modelo como la función de regresión lineal
from sklearn.linear_model import LinearRegression
model= LinearRegression()
```

```python
#Verificamos la función relacionada al modelo 
type(model)
```

```python
#ajustamos el modelo con las variables antes declaradas
model.fit(X=Vars_Indep, y=Var_Dep)
```

```python
#Verificamos los coeficientes obtenidos para el modelo ajustado
model.__dict__
```

```python
#Evaluamos la eficiencia del modelo obtenido por medio del coeficiente R determinado
model.score(Vars_Indep,Var_Dep)
```

#2- Regresión logística

```python
#Declaramos las variables dependientes e independientes para la regresión logística
Vars_Indep= dflimpio[['238_frequency_buying_store','241_age_of_the_interviewee']]
Var_Dep= dflimpio['283_fulfill_needs']
```

```python
#Redefinimos las variables
X= Vars_Indep
y= Var_Dep
```

```python
#Dividimos el conjunto de datos en la parte de entrenamiento y prueba:
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state =None)
```

```python
#Se escalan todos los datos
escalar= StandardScaler()
```

```python
#Para realizar el escalamiento de las variables "X" tanto de entrenamiento como de prueba, utilizaremos fit_transform
X_train = escalar.fit_transform(X_train)
X_test = escalar.transform(X_test)
```

```python
#Definimos el algoritmo a utilizar
from sklearn.linear_model import LogisticRegression
algoritmo = LogisticRegression()
```

```python
#Entrenamos el modelo
algoritmo.fit(X_train, y_train)
```

```python
#Realizamos una predicción
y_pred= algoritmo.predict(X_test)
y_pred 
```

```python
#Verifico la matriz de Confusión
from sklearn.metrics import confusion_matrix
matriz = confusion_matrix(y_test, y_pred)
print('Matriz de Confusión:')
print(matriz)
```

```python
#Calculo la precisión del modelo 
from sklearn.metrics import precision_score
precision= precision_score(y_test, y_pred, average="weighted", pos_label="yes")
print('Precision del modelo:')
print(precision)
```

```python
#Calculo la exactitud del modelo
from sklearn.metrics import accuracy_score
exactitud= accuracy_score(y_test, y_pred)
print('Exactitud del modelo')
print(exactitud)
```

```python
#Calculo la sensibilidad del modelo 
from sklearn.metrics import recall_score
sensibilidad=recall_score(y_test, y_pred, average='weighted', pos_label="yes")
print('Sensibilidad del modelo')
print(sensibilidad)
```
